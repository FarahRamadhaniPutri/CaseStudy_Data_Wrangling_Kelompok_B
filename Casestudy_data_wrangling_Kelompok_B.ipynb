```python
# Code untuk Soal No 01 diberikan sebagai Contoh dan Bonus
# Pelajari dengan baik

nilai_dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1) 
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN', right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]
sb = nilai.groupby(nilai['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_dosen = nilai_dosen.set_index(['NIDN']) 
nilai_dosen = pd.concat([nilai_dosen, sb], axis=1)
nilai_dosen
```

# Interpretasi 01: 

* Dapat terlihat bahwa rata-rata nilai dosen paling tinggi yaitu yang diberikan oleh Andri Basuki = 85.750000, sedangkan jika dibandingkan dengan rata-rata yang diberikan Annisa Fitria = 60.090909. Hal ini menunjukkan adanya perbedaan rata-rata antar dosen yang cukup jauh. 
* Dapat terlihat pula bahwa standar deviasi nilai antar dosen rentangnya cukup jauh, sehingga dapat dikatakan nilai yang diberikan oleh dosen cukup bervariasi. 

Maka, dapat disimpulkan terdapat dosen yang cenderung memberikan nilai lebih tinggi yaitu Andri Basuki atau lebih rendah yaitu Annisa Fitria jika dibandingkan dengan dosen lainnya

```python
# Code Untuk Soal No 02

pendidikan_dosen = pd.merge(dosen, nilai, on='NIDN')
pendidikan_dosen = pendidikan_dosen[['Pendidikan_Terakhir','nilai']]
mean_nilai = pendidikan_dosen.groupby(pendidikan_dosen['Pendidikan_Terakhir']).mean()[['nilai']]
mean_nilai = mean_nilai.rename({'nilai':'mean_nilai'}, axis = 1)
std_nilai = pendidikan_dosen.groupby(pendidikan_dosen['Pendidikan_Terakhir']).std()[['nilai']]
std_nilai = std_nilai.rename({'nilai':'std_nilai'}, axis = 1)
pendidikan_dosen = pd.concat([mean_nilai, std_nilai], axis = 1)
pendidikan_dosen
```

# Interpretasi 02:

* Selanjutnya akan dicari mean (dengan variabel mean_nilai) dan standar deviasidengan variabel std_nilai dari nilai pada tabel pendidikan_dosen dengan menggunakan fungsi grouby
* Kemudian rename kolom 'Nilai' dengan dengan 'mean_nilai' dan 'std_nilai'
* Gabungkan tabel mean_nilai dan std_nilai
* Dapat terlihat bahwa rata-rata nilai yang diberikan dosen dengan pendidikan terakhir S3 lebih tinggi daripada dosen dengan pendidikan terkahir S2 dengan rentang rata-rata nya sekitar 19, hal ini cukup jauh.
* Kesimpulan yang didapat adalah terdapat perbedaan dalam pemberian nilai antara dosen dengan pendidikan terakhir S3 dan S2

```python
# Code Soal No 03

nilai_mahasiswa = nilai.groupby(nilai['nim']).mean()[['nilai']]
nilai_mahasiswa = nilai_mahasiswa.rename({'nilai':'mean_nilai'}, axis=1)
nilai_mahasiswa = pd.merge(mahasiswa, nilai_mahasiswa, left_on='nim', right_index=True)
nilai_mahasiswa = nilai_mahasiswa[['nim', 'Asal_Sekolah', 'mean_nilai']]
sb_m = nilai.groupby(nilai['nim']).std()[['nilai']]
sb_m = sb_m.rename({'nilai':'std_nilai'}, axis=1)
nilai_mahasiswa = nilai_mahasiswa.set_index(['nim'])
nilai_mahasiswa = pd.concat([nilai_mahasiswa, sb_m], axis=1)
nilai_mahasiswa.pivot_table(index='Asal_Sekolah')
```
